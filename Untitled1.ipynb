### 2.2 Рассмотрим в какие дни совершается больше всего заказов

```python
#Посчитаем абсолютные и относительные значения для столбца weekday_key, сгруппировав данные по столбцу order_gk
weekday_key_info =  new_taxi.pivot_table(index = 'weekday_key', columns='driver_response', values='order_gk', aggfunc='count'
                                          ).sort_values(by = 'weekday_key').reset_index()
weekday_key_info.columns = ['weekday','order_accepted', 'order_not_accepted']
weekday_key_info['order_total'] = weekday_key_info['order_not_accepted'] + weekday_key_info['order_accepted']
weekday_key_info['%'] = (weekday_key_info['order_accepted']/weekday_key_info['order_total']*100).round(2)
display(weekday_key_info)
```

```python
#Выбираем размер графика
plt.figure(figsize=(16, 6))
#Построим диаграмму
sns.histplot(
    data = new_taxi,
    hue = 'driver_response',
    x ='weekday_key',
    #stat = 'density',
    common_norm = False,
    #ax = ax2,
    palette = 'light:b'
)
#Формируем заголовок графика и подписи осей
plt.title('Диаграмма распределения дней по количеству заказов')
plt.xlabel('Дни недели')
plt.ylabel('Количество заказов')
#Отображаем график на экране
plt.show()
```

 *Вывод:* Максимальное количество заказов 58% водители принимают в понедельник, а минимальное 44% в четверг.
